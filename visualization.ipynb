```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
df = pd.read_csv('weather.csv')
df.head()
```

```python
df.info()
```

```python
plt.hist(df['Rainfall']) #histogram
```

```python
sns.distplot(df['Sunshine'],hist=False,kde=True,rug=True)
```

```python
sns.kdeplot(df['Rainfall'], shade=True, color="b")#line histogram
```

```python
plt.plot(df['Evaporation'])
```

```python
plt.scatter(df['MinTemp'],df['Temp9am']) #scatter plot two variables
```

```python
plt.scatter(df['MinTemp'],df['MaxTemp'],alpha=0.4, edgecolors='g', linewidth=2) #bubble chart three variables
```

```python
sns.lmplot(y='Pressure3pm',x='Pressure9am',data=df)
```
