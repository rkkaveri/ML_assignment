```python
import re
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/Shreyas3108/Titanic-EDA-and-Survival-prediction/master/train.csv")
df.head()
```

```python
df1 = pd.read_csv("https://raw.githubusercontent.com/Shreyas3108/Titanic-EDA-and-Survival-prediction/master/test.csv")
df1.head()
```

```python
result_df = pd.concat([df,df1])
result_df.head()
```

```python
features = result_df.dropna(subset=["Survived"])
features.head()
```

```python
features.insert(loc=2,column="Has_Cabin",value=features["Cabin"].notna())
features.head()
```

```python
features["title"]=features.Name.apply(lambda x:re.search('([A-Z][a-z]+)\.',x).group(1))
features["title"].unique()
```

```python
features["title"]=features["title"].replace(["Don","Rev","Dr","Major","Lady","Sir","Col","Capt","Countess","Jonkheer"],"special")
features["title"]
```

```python
features["Categorical_age"]=pd.qcut(features.Age,q=4,labels=False)
features["Categorical_fare"]=pd.qcut(features.Fare,q=4,labels=False)
features.head()
```

```python
features = features.fillna(features.median())
features.head()
```

```python
features.isnull().sum()
```

```python
req_features = features.drop(["Cabin","Name","PassengerId","Ticket","Age","Fare","SibSp","Parch"],axis=1)
req_features.head()
```

```python
req_features = pd.get_dummies(req_features)
req_features.head()
```

```python
x = req_features.drop(["Survived"],axis = 1)
y = req_features["Survived"]
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=50)
```

```python
logreg = LogisticRegression()
logreg.fit(x_train,y_train)
```

```python
y_pred = logreg.predict(x_test)
```

```python
result = pd.DataFrame({"Predicted":y_pred,"Actual":y_test})
result.head()
```

```python
logreg.coef_
```

```python
logreg.intercept_
```

```python
logreg.score(x_test,y_test)
```
